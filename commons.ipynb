```python
import json
import time
```

# SAVING FUNCTION

```python
def save_features(file, features):
    t1 = time.time()
    print(f"==============================")
    
    print(f"Saving to {file}...")
    with open(f"commons/{file}", "w") as f:
        json.dump(features, f, indent=4)
        
    t2 = time.time()
    print(f"Finish... Total time: {t2 - t1} seconds")
    print(f"==============================")
    
```
